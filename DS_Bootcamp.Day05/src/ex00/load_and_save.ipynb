# Load and save

Код считывает файл, обрабатывает его и сохраняет результат в файл.

## 1. read_csv

```python
import pandas as pd


input_file = "../data/feed-views.log"
output_file = "../data/feed-views-semicolon.log"

df = pd.read_csv(
    input_file,
    sep='\t',
    header=None,
    names=["datetime", "user"],
    skiprows=[2, 3],
    skipfooter=2,
    engine="python"
    )
```

- Вывод head датафрейма

```python
df.head()
```

- Вывод tail датафрейма

```python
df.tail()
```

- Установка datetime в качестве индексного столбца

```python
df.set_index("datetime", inplace=True)

df
```

## 2. Переименовывание datetimeв в date_time

```python
df.index.name = "date_time"

df
```

## 3. to_csv

```python
df.to_csv(
    output_file, 
    sep=";"
    )
```
