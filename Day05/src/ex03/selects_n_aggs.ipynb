# Selects and aggregations

## 1. Загрузка JSON-файла, созданный в предыдущем упражнении

- Установка столбца CarNumber как индекс.

```python
import pandas as pd


input_file = '../data/auto.json'

df = pd.read_json(input_file)

df.set_index('CarNumber', inplace=True)

df
```

## 2. Выборка

- Строки, где штрафы превышают 2100

```python
df[df["Fines"] > 2100]
```

- Строки, в которых штрафы превышают 2100, а возврат равен 2

```python
df[(df["Fines"] > 2100) & (df['Refund'] == 2)]
```

- Строки, в которых находятся модели из списка: [’Focus’, ’Corolla’]

```python
df[df["Model"].isin(["Focus", "Corolla"])]
```

- Строки, в которых есть номер автомобиля из списка: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]

```python
df.loc[["Y7689C197RUS", "92928M178RUS","7788KT197RUS", "H115YO163RUS", "X758HY197RUS"]]
```

## 3. Выполнение агрегации с помощью make и model

- Медианные штрафы сгруппированные по make

```python
df.groupby(["Make"])["Fines"].median()
```

- Медианные штрафы сгруппированные по make и model

```python
df.groupby(["Make", "Model"])["Fines"].median()
```

- Количество штрафов сгруппированных по make и model

```python
df.groupby(["Make", "Model"])["Fines"].count()
```

- Минимальные и максимальные штрафы, сгруппированные по make и model

```python
max_Fines = df.groupby(["Make", "Model"])["Fines"].max()
min_Fines = df.groupby(["Make", "Model"])["Fines"].min()

max_Fines, min_Fines
```

- Стандартное отклонение штрафов сгруппированных по make и model

```python
df.groupby(["Make", "Model"])["Fines"].std()
```

## 4. Агрегации с номерами автомобилей

- Номера автомобилей сгруппированные по количеству штрафов в порядке убывания

```python
df_Fines_grouped = df.groupby(["CarNumber"])["Fines"].count().sort_values(ascending=False)
df_Fines_grouped
```

- Строки, соответствующие номеру автомобиля, занимающего первое место

```python
df.loc[df_Fines_grouped.index[0]]
```

- Номера автомобилей сгруппированных по сумме штрафов в порядке убывания

```python
sum_Fines = df.groupby(["CarNumber"])["Fines"].sum().sort_values(ascending=False)
sum_Fines
```

- Строки, соответствующие номеру автомобиля, занимающего первое место, из исходного датафрейма

```python
df.loc[sum_Fines.index[0]]
```

- Таблицу, которая отвечает на вопрос: «Существуют ли номера автомобилей, связанные с различными моделями?»

```python
CarNumber_multiModel = df.groupby(df.index)

CarNumber_multiModel.size().sort_values(ascending=False)
```
