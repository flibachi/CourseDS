### Начнем обзор c класса Links

Предварительно отберем из "links.csv" первые 100 строк данных для обработки, в противном случае работа метода будет выполняться чрезмерно долго.

```python
in_f = "ml-latest-small/links.csv"
out_f = "sample_data.csv"
list_of_movies = []
with open(in_f) as input_f, open(out_f, "w") as output_f:
    output_f.write("movieId,imdbId,tmdbId\n")
    lines = input_f.readlines()
    for i in range(1, 100):
        output_f.write(lines[i])
        m_id = lines[i].strip().split(',')[0]
        list_of_movies.append(m_id)
```

```python
l_instance = Links("sample_data.csv")
%timeit Links("sample_data.csv")
```

### Рассмотрим метод, предоставляющий возможность получить список аттрибутов для списка фильмов с сайта imdb.

```python
%%timeit -n 1 -r 1
res = l_instance.get_imdb(list_of_movies, ["director", "RUnTIme", "BUdget", "cumulative worldwide gross"])
print(res)
```
