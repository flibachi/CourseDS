# Select

1. Создать подключения к базе данных с помощью библиотеки sqlite3.

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('../data/checking-logs.sqlite')
```

2. Получить схему таблицы pageviews.

```python
pageviews_schema = pd.read_sql('PRAGMA table_info(pageviews)', conn)

pageviews_schema
```

3. Получить первые 10 строк таблицы pageviews.

```python
pageviews = pd.read_sql('SELECT * FROM pageviews', conn)

pageviews.head(10)
```

4. Получить подтаблицу с помощью одного запроса.

```python
query = """
SELECT uid, datetime
FROM pageviews
WHERE uid > "user_"
ORDER BY 1
"""

pageviews = pd.read_sql(query, conn)

pageviews.set_index("datetime", inplace=True)
pageviews.rename_axis(index={'datetime': 'DatetimeIndex'}, inplace=True)

pageviews.head(10)
```

5. Закрыть соединение с базой данных.

```python
conn.close()
```
